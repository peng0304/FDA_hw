# [Data set](https://www.kaggle.com/lava18/google-play-store-apps)

```python
from collections import Counter 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
data = pd.read_csv('D:\Downloads\jerry\data\googleplaystore.csv')
```

# Top 10 Category count

```python
data_drop = data.drop(columns = ["Genres", "Last Updated", "Current Ver", "Android Ver", "Content Rating"])
data_group = data_drop.groupby(["Category"])
data_count = data_group.size()
data_count.name = "count"
df_data = pd.DataFrame(data_count)
df_one_result = df_data.sort_values(by=['count'], ascending=False)
df_one_result.head(10)
```

# Category count plot

```python
data_category = df_one_result
data_category.plot.bar(color = ["red"],legend = 0)
```

# Most category distribution

```python
most_app = df_one_result.index[0]
most_app_data = data_group.get_group(most_app)
most_app_count = most_app_data.set_index(["Category","Installs"]).count(level="Installs")
most_app_count.plot.bar(color = ["blue"],legend=None)
```

# Category average rate

```python
data_group_mean = data_group.mean()
data_rate_mean = data_group_mean["Rating"]

df_data["Rate mean"] = data_rate_mean
result = df_data.sort_values(by=['Rate mean'], ascending=False)
result.plot.bar(y="Rate mean",legend = None)
```
